## -  Si tuvieras que enseñarme algo en cinco minutos, ¿qué sería y cómo lo harías?

R = Si tuviera que enseñarte algo en cinco minutos, sería sobre mi proyecto personal de Gestión de Restaurantes. Este proyecto, que estoy desarrollando de manera 
independiente, busca optimizar el control de las solicitudes de materia prima en varios restaurantes. La solución que propongo eliminará la necesidad de que los 
administradores se comuniquen individualmente con cada restaurante para verificar productos y cantidades. Actualmente, este proceso se maneja de manera básica 
a través de WhatsApp, pero mi sistema permitirá una gestión centralizada y más eficiente.

Aunque el proyecto aún está en una etapa temprana de desarrollo, ya ha demostrado ser una herramienta valiosa para aprender sobre aspectos cruciales del 
trabajo en equipo y desarrollo.

## -  ¿Cómo gestiona su tiempo y prioriza su día?

proyecto y tarea, lo que me permite mantener una consistencia en mi trabajo y evitar posibles frustraciones. Esta metodología ha resultado en una mejora directa en mi
rendimiento.

En cuanto a la priorización diaria, mi empleo ocupa la mayor parte de mi horario, dedicando aproximadamente 15 horas a la semana a mis proyectos personales y estudios.
Este equilibrio me permite avanzar profesionalmente mientras continúo con mi desarrollo personal y académico.
